```python

# Morphological operations
kernel = np.ones((1, 5), np.uint8)
im6 = cv2.morphologyEx(im4, cv2.MORPH_DILATE, kernel, anchor=(2, 0), iterations=2) 
plt.imshow(im6, cmap="gray")
plt.show()

im7 = cv2.morphologyEx(im6, cv2.MORPH_CLOSE, kernel, anchor=(2, 0), iterations=2)  # 3 instead of 2?
plt.imshow(im7, cmap="gray")
plt.show()

kernel = np.ones((21, 35), np.uint8)
im8 = cv2.morphologyEx(im7, cv2.MORPH_OPEN, kernel, iterations=1)


plt.imshow(im8, cmap="gray")

```

```python

#Find Contours
im9, contours, hierarchy = cv2.findContours(im8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

convex_hull_contours = []
for cnt in contours:
    # get convex hull
    convex_hull_contours.append(cv2.convexHull(cnt))
contours = convex_hull_contours    
```

```python
def crop_minAreaRect(img, rect):

    # rotate img
    angle = rect[2]
    rows,cols = img.shape[0], img.shape[1]
    M = cv2.getRotationMatrix2D((cols/2,rows/2),angle,1)
    img_rot = cv2.warpAffine(img,M,(cols,rows))

    # rotate bounding box
    rect0 = (rect[0], rect[1], 0.0)
    box = cv2.boxPoints(rect)
    pts = np.int0(cv2.transform(np.array([box]), M))[0]    
    pts[pts < 0] = 0

    # crop
    img_crop = img_rot[pts[1][1]:pts[0][1], 
                       pts[1][0]:pts[2][0]]

    return img_crop
```

```python
from scipy.signal import convolve2d
from scipy import ndimage

valid_contours = []
contours_to_merge = []
im_out_cp = im_out.copy()
unscale = 1.0 / scale

if contours != None:
    for contour in contours:
        # If area is not big enough
        if cv2.contourArea(contour) <= 4000:
            continue
        else:
            rect = cv2.minAreaRect(contour)
            #display(rect)
            #y_center = rect[0][1]
            #display(y_center)
            #x_borders = [rect[0][0] + rect[1][0], rect[0][0] + rect[1][1]]
            #x_borders = [rect[0][0] - rect[1][0], rect[0][0] + rect[1][0]]
            #display(x_borders)
            valid_contours.append(contour)

    barcode_images = []      
    for contour in valid_contours:
        #Rectangle of minimum area -> (center_x, center_y), (width, height), angle)
        rect = cv2.minAreaRect(contour)

        bar_code_image = crop_minAreaRect(im, rect)
        if abs(rect[2]) > 45:
            bar_code_image = ndimage.rotate(bar_code_image, 90)
            
        barcode_images.append(bar_code_image)
#         plt.imshow(bar_code_image, cmap='gray')
#         plt.show()
    

        rect = \
            ((int(rect[0][0] * unscale), int(rect[0][1] * unscale)), \
             (int(rect[1][0] * unscale), int(rect[1][1] * unscale)), \
             rect[2])

        # Draw a box around original image
        box = np.int0(cv2.boxPoints(rect))
        cv2.drawContours(im_out_cp, [box], 0, (0, 255, 0), thickness = 2)

plt.imshow(im_out_cp)
plt.show()


```
